```python
import pandas as pd
import numpy as np
import xml.etree.cElementTree as ET
```

```python
tree = ET.ElementTree(file="android-7.1.2_r27.xml")
root = tree.getroot()
```

```python
df = pd.DataFrame()

for child_of_root in root:
    if("project" == child_of_root.tag):
        _df = pd.DataFrame(
            data = [list(child_of_root.attrib.values())],
            columns = child_of_root.attrib.keys())
        df = pd.concat([df, _df], ignore_index=True)
```

```python
df.to_csv("android-7.1.2_r27.csv")
```
